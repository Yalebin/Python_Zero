```python
import basicMod as bsm
import converData as cda
from collections import Counter
import datetime as dt
import numpy as np
import os
import pandas as pd
```

```python
#get station csv file
staFiles = []
for x in os.listdir():
    if '.csv' in x: staFiles.append(x)
#
staFiles
```

```python
staSource = []
for x in staFiles:
    dfTemp = bsm.readStaData(x)
    bsm.cleanSignal(dfTemp)
    dfTemp.drop(index=0, inplace=True)
    dfTemp.reset_index(inplace=True, drop=True)
    #
    print('Now station file:', x)
    bsm.showDfInfo(dfTemp)
    print('*'*80)
    staSource.append(dfTemp)
```

#clean right space for every-Datafram
for x in staSource:
    #clean columns
    bsm.cleanSignal(x)
    #clean cells
    colList = x.columns.to_list()
    bsm.cleanSignal(x, axis=1, columns=colList)

```python
#Transfor item to columns and merge date and houre
staDatas=[]
for x in staSource:
    dfTemp = cda.transDatas(x, '測站', '日期', '測項')
    bsm.showDfInfo(dfTemp)
    print('*'*80)
    staDatas.append(dfTemp)
```
